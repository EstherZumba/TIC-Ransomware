```python
data = pd.read_csv(r"C:\Users\Dell\TIC\Data-Original\Data.csv",low_memory=False)
```

```python
# ==========================================
# FILTRAR TODAS LAS MUESTRAS DE RANSOMWARE Y BENIGNO
# ==========================================

# Normalizar texto para evitar errores por mayúsculas o espacios
data['Category_norm'] = data['Category'].str.strip().str.lower()

# Filtrar únicamente las muestras ransomware y benigno
data_ransom = data[data['Category_norm'].isin(['ransomware', 'benign'])].copy()

# Eliminar la columna auxiliar de normalización
data_ransom = data_ransom.drop(columns=['Category_norm'], errors='ignore')

# Guardar el nuevo dataset con todas las columnas originales
data_ransom.to_csv('dataset_ransomware_vs_benigno.csv', index=False)

print("✅ Dataset ransomware vs benigno guardado correctamente con todas las columnas originales.")
print("Total de muestras:", len(data_ransom))
print(data_ransom['Category'].value_counts())

```
