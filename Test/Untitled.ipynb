```python
import pandas as pd

def dividir_en_bloques(df, tamao=10):
    return [df.iloc[i:i + tamao] for i in range(0, len(df), tamao)]


def main():
    # 1. Cargar el dataset
    ruta_dataset = "test_ransomware.csv"
    df = pd.read_csv(ruta_dataset)

    print(f"Dataset cargado con {len(df)} filas")

    # 2. Dividir en bloques de 10 filas
    bloques = dividir_en_bloques(df, tamao=10)

    print(f"Se crearon {len(bloques)} bloques")

    # 3. Mostrar el contenido de cada bloque
    for i, bloque in enumerate(bloques, start=1):
        print(f"\n--- TestRansomware {i} ---")
        print(bloque)

    # 4. Guardar cada bloque como testransomware_X.csv
    for i, bloque in enumerate(bloques, start=1):
        nombre_archivo = f"testransomware_{i}.csv"
        bloque.to_csv(nombre_archivo, index=False)
        print(f"Archivo guardado: {nombre_archivo}")


if __name__ == "__main__":
    main()
```
